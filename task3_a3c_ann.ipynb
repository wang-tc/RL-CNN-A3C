## Test A3C model in small environment

```python
model = Net(s_dim=15, a_dim=4)
state = torch.load(r'.\task3_a3c_ann\saved_models\a3c_ann_h2w2_intelGhost_lr1e-4_gamma0.9_1mIteration.pt')
model.load_state_dict(state['model_state_dict'])
```

```python
game = PacManJX(maze_row_num=2, maze_column_num=2, maze_row_height=2, maze_column_width=2)
game.blinky_random = False
game.setting.maximum_time = 300
game.agent.net = model
game.policy = A3C_Policy_test
game.random_reset()
game.generate_animation()
ani = HTML(game.animation.to_jshtml())
ani
```

```python
res = 0
for i in range(1000):
    game.policy = A3C_Policy_test
    game.random_reset()
    game.setting.maximum_time = 500
    while not game.process.termination:
        game.run_one_step_without_graph()
#     print('total_reward',game.process.win)
    if game.process.win:
        res+=1
print('The win rate from 1000 sample is: ', res/1000)
```
